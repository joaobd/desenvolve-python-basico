### Q2.

Dado um número inteiro positivo informado pelo usuário, crie um programa em Python que utilize o comando for para calcular e exibir a soma dos números de 1 até o número informado. Ou seja, $1 + 2 + 3 + ... + n$

Exemplo de interação:
```
Digite um número: 5
A soma dos números de 1 a 5 é 15
```


````python
## Escreva e execute seu código aqui
numero = int(input("Digite um número inteiro positivo: "))
soma = 0

for i in range(1, numero + 1):
    soma += i

print(f"A soma dos números de 1 até {numero} é: {soma}")
````

### Q3.

Escreva um programa que lê 10 valores inteiros positivos do usuário e ao final imprime a média dos valores digitados com duas casas decimais.

```
Digite 10 números positivos:
3
2
12
14
20
42
19
0
8
1
A média dos valores digitados é 12.10
```

````python
## Escreva e execute seu código aqui
total_valores = 10
soma = 0

for _ in range(total_valores):
    valor = int(input("Digite um valor inteiro positivo: "))
    if valor < 0:
        print("Por favor, digite apenas valores positivos.")
        break
    soma += valor
else:
    media = soma / total_valores
    print(f"A média dos valores é: {media:.2f}")
````

### Q4.

Você vai criar um sistema que registra os resultados dos jogos do Atlético MG ao longo de um campeonato. Seu sistema vai receber os resultados de todos os jogos do Galo, e deve calcular a pontuação do time sabendo que vitórias valem 3 pontos, empates 1 ponto e derrotas 0 pontos.

Entrada: <br>
A primeira linha de entrada é um inteiro N com a quantidade jogos do galo. Para cada jogo você deve ler 2 inteiros, o primeiro com a quantidade de gols do galo e o segundo com a quantidade de gols do time oponente. 

Saída: <br>
Apresente a soma de vitórias, empates e derrotas do galo, junto com o cálculo da pontuação total.

Ex:
```
Entrada:
4
2
0
1
1
0
3
1
0

Saída:
Vitórias: 2
Empates: 1
Derrotas: 1
Pontuação: 7

```




````python
## Escreva e execute seu código aqui
# Inicializa as variáveis para contagem de vitórias, empates e derrotas
vitorias = 0
empates = 0
derrotas = 0

# Lê a quantidade de jogos do Atlético MG
N = int(input("Digite a quantidade de jogos do Atlético MG: "))

# Loop para ler os resultados de cada jogo
for _ in range(N):
    gols_galo, gols_oponente = map(int, input("Digite os gols do Atlético MG e do time oponente: ").split())

    # Calcula a pontuação com base nos resultados
    if gols_galo > gols_oponente:
        vitorias += 1
    elif gols_galo == gols_oponente:
        empates += 1
    else:
        derrotas += 1

# Calcula a pontuação total
pontos = vitorias * 3 + empates

# Imprime os resultados
print(f"Vitórias: {vitorias}")
print(f"Empates: {empates}")
print(f"Derrotas: {derrotas}")
print(f"Pontuação total: {pontos}")
````

### Q5.

Faça um programa que lê dois inteiros N e M, e imprime na tela um campo de batalha naval. O tabuleiro deve possuir N linhas e M colunas. A primeira linha é composta por um espaço em branco e o cabeçalho das colunas, ou seja, valores de 1 a M. As N linhas seguintes iniciam com o cabeçalho da linha, ou seja, seu número, seguido de M caracteres "/" (barra) indicando uma possível posição jogável. 

```
Entrada:
5
4

Saída:
  1 2 3 4 
1 / / / / 
2 / / / / 
3 / / / / 
4 / / / / 
5 / / / / 
```

Para esse exercício, precisamos lembrar que o comando print implicitamente adiciona uma quebra de linha ao final da impressão. Podemos interferir no final da impressão adicionando mais uma entrada ao print. No exemplo, finalizamos cada linha com um espaço em branco:
```python
print("Texto qualquer", end = " ")
```

````python
## Escreva e execute seu código aqui
# Lê os valores de N e M
N = int(input("Digite o número de linhas (N): "))
M = int(input("Digite o número de colunas (M): "))

# Imprime o cabeçalho das colunas
print("  ", end="")
for coluna in range(1, M + 1):
    print(coluna, end=" ")
print()

# Imprime as linhas do tabuleiro
for linha in range(1, N + 1):
    print(f"{linha} ", end="")
    for _ in range(M):
        print("/", end=" ")
    print()
````

### Q6.

Escreva um programa que lê uma quantidade indefinida de valores e informa o maior e o menor valor digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
12
-1
29
11
-5
0

Saída:
Maior: 29
Menor: -5
```

````python
## Escreva e execute seu código aqui
# Lê os valores até que o usuário digite 0
valor = float(input("Digite um valor (ou 0 para encerrar): "))
maior = menor = valor  # Inicializa com o primeiro valor digitado

while valor != 0:
    valor = float(input("Digite um valor (ou 0 para encerrar): "))
    if valor != 0:
        maior = max(maior, valor)
        menor = min(menor, valor)

# Imprime os resultados
print(f"Maior valor digitado: {maior}")
print(f"Menor valor digitado: {menor}")
````

### Q7.

Escreva um programa que lê uma quantidade indefinida de valores e informa o **produto** dos valores positivos digitados. A leitura é encerrada quando o usuário digitar o valor 0 (zero).

```
Entrada:
2
-1
1
3
-5
0

Saída:
Produto: 6
```

````python
## Escreva e execute seu código aqui
produto_positivos = 1  # Inicializa o produto com 1

while True:
    valor = float(input("Digite um valor (ou 0 para encerrar): "))
    if valor == 0:
        break  # Encerra a leitura quando o usuário digita 0
    elif valor > 0:
        produto_positivos *= valor  # Multiplica o produto pelos valores positivos

if produto_positivos != 1:
    print(f"Produto dos valores positivos: {produto_positivos}")
else:
    print("Nenhum valor positivo foi digitado.")
````

### Q8.

Vamos fazer uma calculadora que aceita expressões aritméticas de qualquer tamanho até que o usuário digite a palavra "Fim". Sua calculadora executa apenas as operações "+", "-".

```
Entrada:
5
+
8
-
3
+
2
Fim

Saída:
12
```

````python
## Escreva e execute seu código aqui
finalizar = False
resultado = 0
operacao = '='

while not finalizar:
    entrada = input("Digite um número: ")
    numero = float(entrada)
    while True:        
            if operacao == "=":
                resultado += numero    
            if operacao == "+":
                resultado += numero
            if operacao == "-":
                resultado -= numero           
            print(f"Resultado parcial: {resultado}")                
            operacao = input("Digite a operação (+ ou -) ou 'Fim' para encerrar: ")
            if operacao.lower() == "fim":
                finalizar = True
                break
            if operacao not in ("+", "-"):
                operacao = '#'
                print("Operação inválida!")
                continue
            break
print(f"Resultado final: {resultado}")
````
